# Pogram

```python
import os
from IPython.display import Audio, display
from paddleocr import PaddleOCR

# Init
ocr = PaddleOCR(use_textline_orientation=True, lang='fr')

img_folder = '/content/'
img = 'useful debate vocabulary_page-0001.jpg'


# Run
result = ocr.ocr(str(img_folder + img))

# Extract
if result and result[0] is not None:
    full_text = "[S1] "
    for line in result[0]:
        text_only = line[1][0] #line[1][0] = only text format (no image, symbole, or schematics)
        full_text += text_only + ". "

    #Save
    with open("input.txt", "w", encoding='utf-8') as f: #utf-8 for french symbols (é, è, à, â, ç)
        f.write(full_text)

    print("Text extracted.")

    # Generate Audio
    os.system("python cli.py --input input.txt --output output.wav") #use the inline command of DIA

    # Play Audio
    if os.path.exists('output.wav'):
        display(Audio('output.wav', autoplay=True))
    else:
        print("Error: Audio was not created.")
else:
    print("No text found.") #or image path is incorrect.
```
